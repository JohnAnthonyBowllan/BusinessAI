### Change DataFrame for Binary Classification Problem

```python
import copy
test_df = copy.deepcopy(RS_DF_1)
```

```python
def myFunc(starValue):
    if starValue == 5:
        newStarValue = 1
    elif starValue == 4:
        newStarValue = 1
    elif starValue == 3:
        newStarValue = 0
    elif starValue == 2:
        newStarValue = 0
    elif starValue == 1:
        newStarValue = 0
    return newStarValue
```

```python
test_df['stars'] = test_df['stars'].apply(myFunc)
test_df.head(10)
```

```python
RS_DF_1 = test_df
RS_DF_1.head(3)
```

### Categorize User Information to Decrease Distance Between Samples within a Class

```python
def categoryOfReviewOfFriendsCount(base, numGroups, count):
    if count == 0:
        return 0
    elif count >= base ** numGroups:
        return numGroups + 1
    else:
        for i in range(numGroups):
            if (count >= base**i) and (count < base**(i+1)):
                return i + 1
```

```python
RS_DF_1['user_num_reviews'] = RS_DF_1['user_num_reviews'].apply(lambda x: categoryOfReviewOfFriendsCount(2,3,x))
```

```python
RS_DF_1['user_num_friends'] = RS_DF_1['user_num_friends'].apply(lambda x: categoryOfReviewOfFriendsCount(2,3,x))
```

```python
RS_DF_1['business_num_reviews'] = RS_DF_1['business_num_reviews'].apply(lambda x: categoryOfReviewOfFriendsCount(2,3,x))
```
