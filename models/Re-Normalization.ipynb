```python
import pandas as pd
import numpy as np
import keras

from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from keras.utils import np_utils
from sklearn import preprocessing
```

```python
test_df_trimmed_encoded = pd.read_csv('neural_df_trimmed_encoded.csv')
```

```python
test_df_trimmed_encoded.head(3)
```

```python
test_df_trimmed_encoded = test_df_trimmed_encoded.drop(test_df_trimmed_encoded.columns[test_df_trimmed_encoded.columns.str.contains('unnamed', case = False)],axis = 1)
test_df_trimmed_encoded.head(4)
```

```python
RS_DF_1_normalized = (test_df_trimmed_encoded - test_df_trimmed_encoded.mean()) / (test_df_trimmed_encoded.max() - test_df_trimmed_encoded.min())
```

```python
RS_DF_1_normalized.head(3)
```

```python
RS_DF_1_normalized.max() - RS_DF_1_normalized.min()
```

```python
RS_DF_1_normalized = RS_DF_1_normalized.sample(frac=1).reset_index(drop=True)
```

```python
RS_DF_1_normalized.head(4)
```

```python
RS_DF_1_normalized = RS_DF_1_normalized.drop(['user_cool', 'user_funny', 'user_userful'], axis=1)
```

```python
RS_DF_1_normalized.head(4)
```

```python
RS_DF_1_normalized.to_csv(path_or_buf='re-normalized.csv')
```
